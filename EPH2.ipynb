####**Aplicaci√≥n de etiquetas comunes**

```
#cambio de etiquetas: G√©nero
dfss1620m2['ESTADO'].replace({1:'Ocupado', 
                    2:'Desocupado'}, inplace=True)
```

```
dfss1620m2['CAT_OCUP'].replace({1:'Patr√≥n', 
                    2:'Cuenta propia',
                    3:'Obrero o empleado',
                    4:'Trabajador familiar sin remuneraci√≥n'}, inplace=True)
```

```
#cambio de etiquetas: G√©nero
dfss1620m2['CH04'].replace({1:'Var√≥n', 
                    2:'Mujer'}, inplace=True)
```

```
#cambio de etiquetas: REGI√ìN
dfss1620m2['REGION'].replace({1:'GBA', 
                    40:'NOA',
                    41:'NEA', 
                    42:'CUYO', 
                    43:'PAMPEANA', 
                    44:'PATAGONIA'}, inplace=True)
```

```
#cambio de etiquetas: EDUCACION
dfss1620m2['NIVEL_ED'].replace({1:'Primario incompleto', 
                    2:'primario completo',
                    3:'Secundario incompleto', 
                    4:'Secundario completo', 
                    5:'Superior universitario incompleto', 
                    6:'Superior universitario completo', 
                    7:'Sin instrucci√≥n',
                    9:'NSNR'}, inplace=True)
```

```
#cambio de etiquetas: AGLOMERADO
dfss1620m2['AGLOMERADO'].replace({2:'Gran La Plata', 
                    3:'Bah√≠a Blanca - Cerri',
                    4:'Gran Rosario',
                    5:'Gran Santa F√©',
                    6:'Gran Paran√°',
                    7:'Posadas',
                    8:'Gran Resistencia',
                    9:'Comodoro Rivadavia - Rada Tilly',
                    10:'Gran Mendoza',
                    12:'Corrientes',
                    13:'Gran C√≥rdoba',
                    14:'Concordia',
                    15:'Formosa',
                    17:'Neuqu√©n ‚Äì Plottier',
                    18:'Santiago del Estero - La Banda',
                    19:'Jujuy-Palpal√°',
                    20:'R√≠o Gallegos',
                    22:'Gran Catamarca',
                    23:'Gran Salta',
                    25:'La Rioja',
                    26:'Gran San Luis',
                    27:'Gran San Juan',
                    29:'Gran Tucum√°n - Taf√≠ Viejo',
                    30:'Santa Rosa ‚Äì Toay',
                    31:'Ushuaia - R√≠o Grande',
                    32:'Ciudad Aut√≥noma de Buenos Aires',
                    33:'Partidos del GBA',
                    34:'Mar del Plata',
                    36:'R√≠o Cuarto',
                    38:'San Nicol√°s ‚Äì Villa Constituci√≥n',
                    91:'Rawson ‚Äì Trelew',
                    93:'Viedma ‚Äì Carmen de Patagones',}, inplace=True)
```

####**Generaci√≥n de cuenta de ocupados/desocupados**

```
dfss1620m2['Ocupados']=np.where(dfss1620m2['ESTADO']=='Ocupado',1,0)
dfss1620m2['Desocupados']=np.where(dfss1620m2['ESTADO']=='Desocupado',1,0)
```
