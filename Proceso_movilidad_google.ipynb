## **Elaboración de tablas**

###Selección de variables que se usarán

```
df1=df.loc[df['country_region_code'].isin(['AR'])] #esto es para seleccionar únicamente los casos locales
```

###Eliminación de variables que no se usarán

```
#soltar variables que no se usarán de la base con información de las provincias
dfaux2.drop(columns=['Gobernador','pob_j_2015','pob_j_2016','pob_j_2017','pob_j_2018','pob_j_2019','pob_j_2020','pob_j_2021','Capital','pob_j_2022','pob_j_2023','pob_j_2024','pob_j_2025'],axis=1,inplace=True)
```

###Normalización

```
#renombrar columnas para obtener la relación de clave interna (columnas en común entre las bases)
df1.rename({'iso_3166_2_code':'Cod ISO 3166 alpha 2'},axis=1,inplace=True)
```

```
df1['sub_region_2'].replace({' Partido':'',' Department':''},regex=True,inplace=True)
```

```
df1['sub_region_1'].replace({' Province':''},regex=True,inplace=True)
```

```
df1['sub_region_2']=df1['sub_region_2'].astype(str)
dfaux['sub_region_2']=dfaux['sub_region_2'].astype(str)
```

```
df1['sub_region_2'].replace({'9 de Julio':'Nueve de Julio'},regex=True,inplace=True)
```

```
#unión de tablas con descripciones normalizadas e info adicional
df1p = pd.merge(df1, 
               dfaux2, 
               on ='Cod ISO 3166 alpha 2', 
               how ='left')
```

```
df2p = pd.merge(df1p, 
               dfaux, 
               on ='sub_region_2', 
               how ='left')
```

```
df2p.head(10)
```

```
df2p['lat']=df2p['lat'].astype(str)
df2p['long']=df2p['long'].astype(str)
df2p['lat.long']=df2p['lat']+','+df2p['long']
```

```
df2p.drop(columns=['lat','long','place_id'],axis=1,inplace=True)
```
