## **Levantar bases**

```
#levantar desde enlace web
url='https://covid.ourworldindata.org/data/owid-covid-data.csv'
dfint= pd.read_csv(url)
```

###**Base de datos de vacunación**

```
#levantar base sobre vacunas desde enlace web en github de owid
url='https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/vaccinations.csv'
dfintvacc=pd.read_csv(url)
```

## **Procesos**

###**Formato**

```
#para pasar del inglés al castellano las etiquetas de los países países
dfint['location'].replace({'United States':'EEUU', 
                    'World':'Mundo',
                    'Spain':'España', 
                    'United Kingdom':'Reino Unido',
                    'Italy':'Italia', 
                    'Brazil':'Brasil'}, inplace=True)
```

```
dfintvacc['location'].replace({'United States':'EEUU', 
                    'World':'Mundo',
                    'Spain':'España', 
                    'United Kingdom':'Reino Unido',
                    'Italy':'Italia', 
                    'Brazil':'Brasil'}, inplace=True)
```

### **Elaboración de tabla resumen**

```
#resumen tabla internacional seleccionando algunos países (EEUU, Italia, Brasil, España, Argentina, Chile, Peru y "El Mundo")
dfintfiltrada=dfint.loc[dfint['location'].isin(['Mundo','EEUU','Brasil','Argentina','Italia','España','Chile','Uruguay','Paraguay','India','Reino Unido','Peru'])] #esto es para seleccionar únicamente algunos países y el mundo
```

```
dfintvaccfiltrada=dfintvacc.loc[dfintvacc['location'].isin(['Mundo','EEUU','Brasil','Argentina','Italia','España','Chile','Uruguay','Paraguay','India','Reino Unido','Peru'])] #esto es para seleccionar únicamente algunos países y el mundo
```

###**Unión**

```
#unión de datos de contagios y de vacunas por país
dffinal = pd.concat([dfintfiltrada,dfintvaccfiltrada],sort=False) #por defecto es un "outer join"
dffinal
```
